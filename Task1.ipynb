# Handle missing values by filling with the median

```python
df.fillna(df.median(numeric_only=True), inplace=True)

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Verify changes
print("\nMissing Values After Cleaning:\n", df.isnull().sum())
print("\nDataset Shape After Cleaning:", df.shape)
```

```python
import pandas as pd

df = pd.read_csv("extracted_folder/student-mat.csv", delimiter=";")  # Specify delimiter
print(df.columns)  # Check column names
```

```python
print(df['G3'].mean())  # Should work without KeyError
```

```python
avg_G3 = df['G3'].mean()
print("Average final grade (G3):", round(avg_G3, 2))
```

```python
students_above_15 = df[df['G3'] > 15].shape[0]
print("Number of students scoring above 15 in G3:", students_above_15)
```

```python
correlation = df[['studytime', 'G3']].corr()
print("\nCorrelation between study time and G3:\n", correlation)

# Visualizing correlation
sns.scatterplot(x=df['studytime'], y=df['G3'])
plt.title('Study Time vs. Final Grade')
plt.xlabel('Study Time (Hours)')
plt.ylabel('Final Grade (G3)')
plt.show()
```

```python
avg_G3_by_gender = df.groupby('sex')['G3'].mean()
print("\nAverage final grade by gender:\n", avg_G3_by_gender)
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['G3'], bins=10, kde=True, color='blue')
plt.title('Distribution of Final Grades (G3)')
plt.xlabel('Final Grade (G3)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['studytime'], y=df['G3'], color='green')
plt.title('Study Time vs. Final Grade (G3)')
plt.xlabel('Study Time (Hours per Week)')
plt.ylabel('Final Grade (G3)')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
avg_G3_by_gender = df.groupby('sex')['G3'].mean().reset_index()

sns.barplot(x='sex', y='G3', data=avg_G3_by_gender, palette="coolwarm")
plt.title('Average Final Grade (G3) by Gender')
plt.xlabel('Gender')
plt.ylabel('Average G3')
plt.show()
```
